# Density and Contour Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

## Visualizing a 3D Function

```python
# Define a function for 
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# Generate the x and y values to be plotted
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

# Generate a meshgrid to plot the contours
X, Y = np.meshgrid(x,y)

# Generate the Z values for the meshgrid
Z = f(X,Y)

# Generate the contour plot
plt.contour(X, Y, Z, colors='k')
```

```python
# You can add a color map to the contours to show relief in color
plt.contour(X, Y, Z, 20, cmap='RdBu')
```

```python
# a filled contour map can also be generated showing relief with color
plt.contourf(X, Y, Z, 20, cmap='RdBu')
plt.colorbar();
```

```python
# A better way to display the graph is to use

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdBu', interpolation='lanczos')
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, 5, colors='k') # Plot a contour map
plt.clabel(contours, inline=True, fontsize=8)  # Draw the contour labels

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdBu', interpolation='hamming', alpha=.7)
plt.colorbar();
```
