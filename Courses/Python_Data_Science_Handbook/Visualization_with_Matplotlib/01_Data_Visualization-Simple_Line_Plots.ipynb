# Simple Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# The first step is to create a figure and axes

fig = plt.figure()  # Contains all of the data about the figure

ax = plt.axes()     # The data about the bounding box tick marks and 
```

```python
# After creating the figure and axis you can use the 
# plot method on the ax object to plot the (x,y) values on the figure

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

```python
# The pylab interface can also be be used.  This creates the figure and axes in the background.

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# Use the linestyle and color keywords to modify the appearance of the graph.
# You can also use a single non keywork argument.

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), ':k')
```

## Adjusting the Plot : Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5,1.5)
```

```python
# If you want the axes reversed you can reverse the order of the limits

plt.plot(x, np.sin(x))
plt.xlim(11, -1)
plt.ylim(1.5,-1.5)
```

```python
# Limits can also be set with a single call to the plt.axis() method

plt.plot(x, np.sin(x))
plt.axis([0, 2*np.pi, -1.5, 1.5])
```

```python
# plt.axis() allows you to use special strings to set the axes to automatic settings.
#
# 'tight' sets the upper and lower bounds of the axes to the max and min of the figure in each axis.

plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
# use plt.axis() with 'equal; so that the x and y scales are equal.

plt.plot(x, np.sin(x))
plt.axis('equal')
```

## Labeling Plots

```python
# Include Plot Title and axis labels
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
# Include Plot Title, Axis labels, and a legend
plt.plot(x, np.sin(x), '--g', label='sin(x)')
plt.plot(x, np.cos(x), ':k', label='cos(x)')
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.axis('equal')

plt.legend();

```

```python
# With the object oriented interface you can use the 'ax.set()' method to adjust numerous axis
# settings at once.

ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot');
```
