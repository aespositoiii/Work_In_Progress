## 'show( )' or No 'show( )'?   How to display your plots.

```python
# Plotting from a Script
#
# When plotting within a script use plt.show() to display the plot in the terminal
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()

```

```python
# From within a notebook
#
# '%matplotlib notebook' creates an interactive plot
# '%matplotlib inline' creates a static plot

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

## Two Interfaces for the price of one

```python
# MATLAB-style interface

plt. figure() # creates a plot figure

# Create the first of two panels and set current axis
plt.subplot(2,1,1) # (Rows, Columns, Panel Number)
plt.plot(x, np.sin(x))

# Create the second panel and set current axis
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
# Object Oriented Interface
#
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate objects
ax[0].plot(x, np.sin(x), 'orange')
ax[1].plot(x, np.cos(x), 'green');
```
