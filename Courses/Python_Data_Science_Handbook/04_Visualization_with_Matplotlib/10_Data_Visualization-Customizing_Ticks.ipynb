# Customizing Ticks

## Major and Minor Ticks

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

```python
X = np.linspace(0,1000,10000)
Y = X**2
ax = plt.axes(xscale='log', yscale='log')
plt.plot(X,Y)
ax.grid;
```

## Hiding Ticks or Labels

```python
ax = plt.axes()
ax.plot(np.random.rand(50))

ax.yaxis.set_major_locator(plt.NullLocator())      # sets the labels and ticks to null
ax.xaxis.set_major_formatter(plt.NullFormatter())  # sets only the labels to null
```

```python
# Get some face data from scikit-learn
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images
```

```python
# sometimes it makes sense to hide ticks and labels such as when plotting a set of images.

fig, ax = plt.subplots(5, 5, figsize=(7, 7))
fig.subplots_adjust(hspace=.2, wspace=.2)

for i in range(5):
    for j in range(5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], cmap="bone")
```

```python
# sometimes it makes sense to hide ticks and labels such as when plotting a set of images.

fig, ax = plt.subplots(5, 5, figsize=(7, 7))
fig.subplots_adjust(hspace=0, wspace=0)

for i in range(5):
    for j in range(5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], cmap="bone")
```

## Reducing or Increasing the Number of Ticks

```python
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True)
```

```python
# Set the maximum number of ticks allowed in a plot

for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(plt.MaxNLocator(5))
fig
```

## Fancy Tick Formats

```python
# plot a sine and a cosine wave.

fig, ax = plt.subplots()
x = np.linspace(0, 3*np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x),':r', lw=3 , label='Cosine')

# Set up grid, legend and limits

ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, x.max())
```

```python
# Set new tick locations

ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 3))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 6))
fig
```

```python
# Format the axis labels to indicate multiples of pi

def format_func(value, tick_number):
    # find number of multiples of pi/3
    N = int(np.round(3 * value / np.pi))
    if N == 0:
        return "0"
    if N == 1:
        return r"$\pi/3$"
    if N == 2:
        return r"$2\pi/3$"
    if N == 3:
        return r"$\pi$"
    if N % 3 > 0:
        return r"${0}\pi/3$".format(N)
    else:
        return r"${0}\pi$".format(N//3)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```
