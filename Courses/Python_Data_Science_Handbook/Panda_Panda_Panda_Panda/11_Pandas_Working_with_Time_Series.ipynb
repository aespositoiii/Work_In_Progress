## Resampling, Shifting, and Windowing

```python
from pandas_datareader import data
```

```python
goog = data.DataReader('GOOG', start='2004', end='2016', data_source='yahoo')
goog.head()
```

```python
googCL = goog['Close']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
goog.Close.plot()
```

```python
goog.iloc[1000:1500,0:4].plot()
```

## Resampling and Converting frequencies

```python
goog['Close'].plot(alpha = .5, style='-')
goog['Close'].resample('3Y').mean().plot(style=':')
goog['Close'].asfreq('BA').plot(style='--');
plt.legend(['input','resample','asfreq'],loc='upper left')
```

```python


fig, ax = plt.subplots(2, sharex=True)
data = goog.iloc[:10]

data['Close'].asfreq('D').plot(ax=ax[0], marker='o')

data['Close'].asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data['Close'].asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
ax[1].legend(["back-fill", "forward-fill"]);

```

## Time-Shifts

```python
fig, ax = plt.subplots(3, sharey = True)

# Apply a fequency
goog = goog['Close'].asfreq('D', method='pad')

goog.plot(ax=ax[0])
goog.shift(900).plot(ax=ax[1])   # Shifts the data by 900 days
goog.tshift(900).plot(ax=ax[2])  # Shifts the Index by 900 days ***DEPRECATED***

# Legends and Annotations
local_max = pd.to_datetime('2007-11-05')
offset = pd.Timedelta(900, 'D')

ax[0].legent(['input'], loc=2)
ax[0].get_xticklabels()[2].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=.03, color='red')

ax[1].legent(['shift(900)'], loc=2)
ax[1].get_xticklabels()[2].set(weight='heavy', color='red')
ax[1].axvline(local_max, alpha=.03, color='red')


ax[0].legent(['tshift(900)'], loc=2)
ax[0].get_xticklabels()[1].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=.03, color='red')
```

```python
ROI = 100 * (goog.shift(-365) / goog-1)
ROI.plot()
```

```python
rolling = goog.rolling(365, center=True)
data = pd.DataFrame({'input': goog,
                     'one-year rolling_mean': rolling.mean(),
                     'one-year rolling_std': rolling.std()})
ax = data.plot(style=['-','--',':'])
ax.lines[0].set_alpha(.3)
```

## Example : Visualizing Seattle Bicycle Counts

```python
data = pd.read_csv('/Users/anthonyesposito/FremontBridge.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data.shape
```

```python
data.columns = ['Total','West', 'East']
data.head()
```

```python
data.dropna().describe()
```

```python
data.plot()
plt.ylabel('Hourly Bicycle Count')
```

```python
weekly = data.resample('Y').sum()
weekly.plot(style=[':','--','-'])
plt.ylabel('Yearly Bicycle Count (Mil)')
```

```python
weekly = data.resample('W').sum()
weekly.plot(style=[':','--','-'])
plt.ylabel('Weekly Bicycle Count')
```

```python
weekly = data.resample('M').sum()
weekly.plot(style=[':','--','-'])
plt.ylabel('Monthly Bicycle Count')
```

```python
daily = data.resample('D').sum()
daily['2015':'2017'].rolling(30,center=True).sum().plot(style=[':','--','-'])
plt.ylabel('mean hourly count')
```

```python
daily = data.resample('d').sum()
daily.loc['2016':'2018'].rolling(30,center=True).sum().plot(style=[':','--','-'])
plt.ylabel('mean monthly count')
```

```python
daily.loc['2015'].rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);
```

### Digging Into the data

```python
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':','--','-'])
```

```python
by_dayOFweek = data.groupby(data.index.dayofweek).mean()
by_dayOFweek.index=['Mon', 'Tues', 'Weds', 'Thurs', 'Fri', 'Sat', 'Sun']
by_dayOFweek.plot(style=[':','--','-'])
```

```python
by_dayOFyear = data.groupby(data.index.dayofyear).mean()

by_dayOFyear.plot(style=[':','--','-'])
```

```python
by_dayOFyear = data.groupby(data.index.dayofyear).mean()

by_dayOFyear.rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);
```

```python
weekend = np.where(data.index.weekday <5, 'Weekday', 'Weekend')
by_time = data.groupby([weekend, data.index.time]).mean()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2, figsize=(14,5))
by_time.loc['Weekday'].plot(ax=ax[0], title = 'Weekdays',
                           xticks = hourly_ticks, style=[':', '--', '-'])
by_time.loc['Weekend'].plot(ax=ax[1], title = 'Weekends',
                           xticks = hourly_ticks, style=[':', '--', '-'])
```
