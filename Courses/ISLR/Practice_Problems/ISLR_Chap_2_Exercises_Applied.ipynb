    * Room.Board : Room and board costs
    * Books : Estimated book costs
    * Personal : Estimated personal spending
    * PhD : percent of faculty with Ph.Ds
    * Terminal : Percent of faculty with terminal degree
    * S.F.Ratio : Student/Faculty ratio
    * perc.alumni : percetn of alumni who donate
    * Expend : Instructional expenditure per student
    * Grad.Rate : Graduation Rate

```python
data = pd.read_csv('/Users/anthonyesposito/Desktop/GitWork/Work_In_Progress/Courses/ISLR/ISLR_DATA/College.csv')
data.head()
```

```python
data.rename(columns={'Unnamed: 0': 'College'}, inplace=True)
data.head()
```

```python
data.rename(columns={'F.Undergrad':'F_Undergrad', 'P.Undergrad':'P_Undergrad', 'Room.Board':'Room_Board', 'S.F.Ratio':'StoF_Ratio', 'Grad.Rate':'Grad_Rate', 'per.alumni':'perc_alumni' }, inplace=True)
```

```python
data.head()
```

```python
data.set_index('College', inplace=True)
data.head()
```

```python
data.query('Top10perc > 80')
```

```python
sns.pairplot(data.iloc[:100, 0:8])
```

```python
data['AccRate'] = data.eval('Accept / Apps')
```

```python
data['PTtoFT_Ratio'] = data.eval('P_Undergrad / F_Undergrad')
```

```python
data['EnrollRate'] = data.eval('Enroll / Accept')
```

```python
sns.pairplot(data.loc[:,['EnrollRate','PTtoFT_Ratio','AccRate', 'Top10perc']])
```

```python
Elite = data.query('(Top10perc > 50) | (Top25perc > 90)')
Private = data.query('Private == "Yes"')
Public = data.query('Private != "Yes"')
PublicElite = data.query('(Private != "Yes") & ((Top10perc > 50) | (Top25perc > 90))')
```

```python
fig, ax = plt.subplots()
ax.set_title('Multiple Samples with Different sizes')
ax.boxplot([data.loc[:,'AccRate'], Private.loc[:,'AccRate'], Public.loc[:,'AccRate'], PublicElite.loc[:,'AccRate'], Elite.loc[:,'AccRate']])

plt.show()
```
