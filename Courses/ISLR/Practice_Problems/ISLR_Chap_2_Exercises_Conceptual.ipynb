1. What are the advantages and disadvantages of a very flexibl (versus a less flexible) approach for regression or classificatino? Under what circumstances might a more flexible approach be preferred to a less flexible approach? hwen might a less flexible approach be preferred?
    * Advantages of a very flexible model:
        * Can be accurate for highly irregular data
        * Can utilize a small training sample size when each sample has a large number of parameters
    * Disadvantages
        * It is likely that the model will not handle new data well if it is significantly different from the training data 
    * A less flexible approach may be more favorable with a large sample size with a small set of parameters for each sample

1. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification ( as opposed to a non-parametric approach)? 
    * Parametric approach takes a mathematical form and developes coefficients for that form based on each of the parameters measured in the data.
    * Non-parametric approaches simply develops a model without any assumption of the form
    * If there is a reasonable degree of certainty in the form of the data (ie. linear, quadratic, logarithmic etc.) a parametric approach is suitable.  The number of samples required to develop a parametric model is much less than a non-parametric approach since the assumption of form restricts the variability of the model.
