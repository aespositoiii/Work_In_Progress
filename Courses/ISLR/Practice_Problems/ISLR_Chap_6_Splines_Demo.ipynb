```python
def SplineFeat(x, degree=3, n_knots=3, knots=0):
    x = x[:, np.newaxis]
    xReg = x
    for i in range(1, degree):
        xReg = np.append(xReg, x**(i+1), axis=1)
        
    if isinstance(knots, int):
        space = ( x.max() - x.min()) / ( n_knots + 1)
        knots = np.array([])
        for i in range(n_knots):
            knots = np.append(knots, x.min() + space *( i + 1 ))
    
    for i in knots:
        
        xtemp = x - i
        xtemp[ xtemp < 0 ] = 0
        xReg = np.append(xReg, xtemp**3, axis=1)
            
    
    return xReg, knots
```
