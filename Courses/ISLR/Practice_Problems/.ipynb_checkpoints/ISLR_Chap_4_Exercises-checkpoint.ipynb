```python
autos[['Make', 'Model']]= autos['name'].str.split(' ', 1, expand=True)
autos = autos.drop('name', axis=1)
```

```python
autos['mpgHiLo'] = np.nan
autos.head()
```

```python
for i in range(autos.shape[0]):
    if autos.iloc[i, 0] > autos['mpg'].median():
        autos.iloc[i, 8] = 'Hi'
    else:
        autos.iloc[i, 8] = 'Lo'
```

```python
autos.tail(20)
```

```python
X = autos.drop('mpgHiLo', axis=1)
y = autos['mpgHiLo']
y.unique()
```

```python
print(X.head(20))
print(y.head(20))
```

```python
model1 = LogisticRegression()
scores1 = cross_validate(model1, X[['displacement', 'horsepower', 'weight', 'acceleration']], y)
scores1['test_score'].mean()
```

```python
model2 = LinearDiscriminantAnalysis()
scores2 = cross_validate(model2, X[['displacement', 'horsepower', 'weight', 'acceleration']], y)
scores2['test_score'].mean()
```

```python
model3 = QuadraticDiscriminantAnalysis()
scores3 = cross_validate(model3, X[['displacement', 'horsepower', 'weight', 'acceleration']], y)
scores3['test_score'].mean()
```

```python
from sklearn.model_selection import validation_curve

K = np.arange(1, 100)
train_score, val_score = validation_curve(KNeighborsClassifier(), X, y, param_name='n_neighbors', param_range=K, cv=5)

plt.plot(K, np.mean(train_score, 1), color='b', label = 'training score')
plt.plot(K, np.mean(val_score, 1), color='r', label = 'validation score')
plt.legend(loc='best')
plt.ylim(0, 1)
plt.xlabel('K')
plt.ylabel('score');
OptimumK = (K[(np.mean(val_score, 1) == np.mean(val_score, 1).max())])
print('Optimum K       : ', OptimumK)
print('Optimum K Score : ', np.mean(val_score, 1)[K==OptimumK])
```
